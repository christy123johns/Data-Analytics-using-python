```python
# create new df_col dataframe from df.copy() method. 
df_col = df.copy()
```

```python
# change float64 to float32 of Referal columns
df_col['Amount_spent'] = df_col['Amount_spent'].astype('float32')
```

```python
# Object datatype to category convert

df_col['Gender'] = df_col['Gender'].astype('category')
df_col['Segment'] = df_col['Segment'].astype('category')
df_col['Marital_status'] = df_col['Marital_status'].astype('category')
df_col['Employees_status'] = df_col['Employees_status'].astype('category')
df_col['Payment_method'] = df_col['Payment_method'].astype('category')
```
