# Multiple linear regression

```python
df = pd.read_excel('M_Regression.xlsx')
df
```

```python
import statsmodels.formula.api as smf
formula="Price~Area+Bedrooms+Age_house "
reg1=smf.ols(formula,df).fit()
print(reg1.summary())
```

# another way to find answer

```python
dfq = pd.read_excel('M_Regression.xlsx')
dfq
```

```python
X = dfq[['Area', 'Bedrooms', 'Age_house']]
X
```

```python
y = dfq['Price']
y
```

```python
reg = linear_model.LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
 reg.predict([[1700, 4, 15]])
```

## Data preprocessing (null values by median value)

```python
df1 = pd.read_excel('M_Regression_2.xlsx')
df1
```

```python
df1.Bedrooms.median()
```

```python
df1.Bedrooms = df1.Bedrooms.fillna(df1.Bedrooms.median())
df1
```

```python
import statsmodels.formula.api as smf
formula="Q('Price')~Area+Bedrooms+Age_house "
reg1=smf.ols(formula,df1).fit()
print(reg1.summary())
```
