# T Test - 2 sample t test / Independent sample t test

The Independent Samples t Test or 2-sample t-test compares the means of two independent groups in order to determine whether there is statistical evidence that the associated population means are significantly different. The Independent Samples t Test is a parametric test. This test is also known as: Independent t Test

1. This test is a Parametric test
2. If the groups come from two different populations
3. we use T-test when the population standard deviation is unknown (not given) and the sample size is small (<30)
4. T-test is also known as small sample test.
5. It is a statistical technique that is used to analyze the mean comparison of two independent groups.

e.g., testing the performance score of two batch.(Batch 1 and 2)

# For example
The required score on a final exam is 70. The test is run to determine which batch of trainees performs best. Twenty pupils are chosen at random to measure effectiveness Hence, random scores from two batches are collected and analyzed. 
Which batch has more productive students?



Ho : Means of two groups of a Sample are Equal
    
H1 : Means of two groups of a Sample are not equal

# 2- Sample t test using training efficiency data

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import ttest_rel
```

```python
df=pd.read_excel('Training Efficiency.xlsx')
```

```python
df
```

```python
df.columns
```

```python
df['Gender'].unique()
```

## Hypothesis

Hypothesis stated for this analysis:

Ho : Means of two groups(Male and Female) of a Sample are Equal
    
H1 : Means of two groups of a Sample are not equal


```python
#gender have only 2 groups-Male and Female
group1 = df[df['Gender']=='Male']
group2 = df[df['Gender']=='Female']
```

```python
group1
```

```python
group2
```

```python
df[['exam score']].groupby(df['Gender']).mean()
```

```python
ttest, pvalue = ttest_ind(group1['exam score'], group2['exam score'])
ttest, pvalue
```

```python
print(pvalue)
```

```python
# alpha value is 0.05 or 5% of significance level

if pvalue > 0.05:
    print("Accept Null Hypothesis (Ho: means of two groups are equal)")
else:
    print("Reject Null Hypothesis (Ho) & Accept (H1)")
```
