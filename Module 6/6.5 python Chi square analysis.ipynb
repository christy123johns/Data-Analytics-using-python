# Chi Square Analysis

The test is applied when you have two categorical variables from a single population. It is used to determine whether there is a significant association between the two categorical variables.

1. It is a non parametric test
2. used for testsing goodness of fit
3. test independence of two  variables
4. chi square test tests the significance of association between 2 variables

Chi square test is a test of statistical significance for categorical variables.it is used to test the statistical significance of the observed frequency with respect to the expected frequency.

To calculate chi- square, we take the square of differences between observed and expected values and divide it by the expected values

Hypothesis stated:
    
    Ho: There is No significant difference between 2 categorical variables
    H1: There is Significant difference between 2 categorical variables


```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.stats import chi2_contingency
```

```python
df=pd.read_excel('Training Efficiency.xlsx')
df
```

```python
ctab1=pd.crosstab(df['Gender'],df['Status'])
ctab1
```

```python
import matplotlib.pyplot as plt
import  scipy.stats as stats
from scipy.stats import chi2

ctab1.plot(kind='bar', cmap ='cividis')
plt.title('Gender vs Final status\n', fontdict= {'fontsize':20, 'fontweight':5, 'color':'darkblue'})
```

```python
alpha=0.05
values=stats.chi2_contingency(ctab1)
values
```

```python
p_value1=values[1]
p_value1
```

```python

if p_value1 <= alpha:
    print('reject H0 ie, there is a relation between 2 variables')
else:
    print('accept H0 ie, there is no relation between 2 variables ')
```

# Using Gender and batch

```python
ctab2=pd.crosstab(df['Gender'],df['Batch'])
ctab2
```

```python
ctab2.plot(kind='bar', cmap ='coolwarm')
plt.title('Gender vs Batch\n', fontdict= {'fontsize':20, 'fontweight':5, 'color':'darkblue'})
```

```python
alpha=0.05
values_2=stats.chi2_contingency(ctab2)
values_2
```

```python
p_value2=values_2[1]
p_value2
```

```python

if p_value2 <= alpha:
    print('reject H0 ie, there is a relation between 2 variables')
else:
    print('accept H0 ie, there is no relation between 2 variables ')
```
