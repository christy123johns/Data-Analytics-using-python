# Anova Test

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

It tell whether two or more groups are similar or not based on their mean similarity and f-score.

Example : there are 3 different Batch and their exam score need to check whether all 3 group are similar or not

Hypothesis stated:
    
    Ho = Means are not different in each batch
    H1 = means are differnt is each batch

```python
df=pd.read_excel('Training Efficiency.xlsx')
df
```

```python
df_anova=df[['Batch','exam score']]
df_anova
```

```python
df['Batch'].unique()

#or
#grps = pd.unique(df_anova.Batch.values)
#grps
```

```python
df_anova['exam score']
```

```python
batch1 = df[df['Batch']=='Batch 1']
batch2 = df[df['Batch']=='Batch 2']
batch3 = df[df['Batch']=='Batch 3']
```

```python
batch1
```

Hypothesis stated:
    
    Ho = Means are not different in each batch
    H1 = means are differnt is each batch

```python
f,p=stats.f_oneway(batch1['exam score'],batch2['exam score'],batch3['exam score'])
f,p
```

```python
print(p)
```

```python
if p>0.05:
    print("accept null hypothesis")
else:
    print("reject null hypothesis")
```
