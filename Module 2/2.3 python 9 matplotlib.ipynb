# Introduction to matplotlib package

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

xpoints = np.array([0, 2]) #(x1,x2)
ypoints = np.array([0, 100])  #(y1,y2)

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([4, 12])   #(x1,x2)
ypoints = np.array([2, 8])    #(y1,y2)

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([2, 7])
ypoints = np.array([3, 11])

plt.plot(xpoints, ypoints, marker='o')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, marker='*')
plt.show()
```

```python
ypoints = np.array([2, 8, 1, 11, 5, 9])

plt.plot(ypoints, marker='+')
plt.show()
```

```python
ypoints = np.array([2, 8, 1, 12])

plt.plot(ypoints, marker = 'o')
plt.show()
```

```python
productCategory = np.array(['Furniture', 'Technology', 'Office Supplies'])
sales = np.array ([4110451.90, 4744557.50, 3787492.52] )

plt.bar(productCategory,sales)
plt.show()
     
```
