```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read excel file 

df = pd.read_excel('Online_Store_customer_data.xlsx')
df.head(3)
```

```python
sns.barplot(df['Payment_method'],df['Amount_spent'])
plt.show()
```

```python
sns.barplot(df['Payment_method'],df['Amount_spent'], color='yellow')
plt.show()
```

```python
sns.countplot(x= 'Payment_method',data= df, palette='Blues')
```

```python
sns.countplot(x= 'Payment_method',hue= 'Gender',data= df, palette='rocket')
```

```python
sns.boxplot(x='Payment_method',y='Amount_spent', data=df)
```
