# To Check Normal Distribution

```python
import numpy as np
import pandas as pd
```

```python
# read excel file 

df = pd.read_excel('Online_Store_customer_data.xlsx')
df.head(3)
```

### Histogram

```python
df.plot(
    y='Age',
    kind='hist',
    bins=10)
```

### KDE plot

```python
df.plot(
    y='Age',
    xlim=(0, 100),
    kind='kde')
```
