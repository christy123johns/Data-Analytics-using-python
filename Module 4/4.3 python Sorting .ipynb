# Sortting

```python
import numpy as np
import pandas as pd
```

```python
# read file

df = pd.read_excel('Online_Store_customer_data.xlsx')
df.head(4)
```

```python
df.shape
```

```python
# Sort Values by State_names
df.sort_values(by=['State_names']).head(3)
```

```python
df.sort_values(by=['Amount_spent']).head(2)
```

```python
# Sort Values Amount_spent with ascending order
df.sort_values(by=['Amount_spent']).tail(3)
```

```python
# Sort Values Amount_spent with descending order
df.sort_values(by=['Amount_spent'], ascending=False).head(3)
```

Alternatively, We can use nlargest() and nsmallest() functions, 
for displaying largest and smallest values with desired numbers. 
for example, If we want to display 4 largest Amount_spent rows then we use this:

```python
# nlargest
df.nlargest(4, 'Amount_spent').head() 
# first argument is how many rows you want to disply and second one is columns name
```

```python
# nsmallest
df.nsmallest(3, 'Age').head()
```
