# Grouping

```python
import numpy as np
import pandas as pd
```

```python
# read file

df = pd.read_excel('Online_Store_customer_data.xlsx')
df.head(4)
```

# Summarizing or grouping data
Groupby Grouping by one column: For example, if we want to find maximum values of Age and Amount_spent by Gender then we can use this:

```python
df[['Age']].groupby(df['Gender']).max()
```

```python
df[['Age']].groupby(df['Gender']).min()
```

```python
#group the data to view maximum Age and Maximum Amount spent in each gender seperately

df[['Age', 'Amount_spent']].groupby(df['Gender']).max()
```

To find mean, count, and max values of Age and Amount_spent by Gender then we can use agg() function with groupby()

```python
# Group the data to get informations like mean, count and maximum values of Age and Amount spent in each gender seperately

state_gender_res = df[['Age','Gender','Amount_spent']].groupby(['Gender']).agg(['count', 'mean', 'max'])
state_gender_res
```

```python
# Group by one columns

state_gender_res = df[['Age','Amount_spent']].groupby(df['Gender']).agg(['count', 'mean', 'max'])
state_gender_res
```

Grouping by multiple columns: To find total count, maximum and minimum values of Amount_spent by State_names, Gender, and Payment_method then we can pass these columns names under groupby() function and add .agg() with count, mean, max argument.

```python
#Group By multiple columns

state_gender_res = df[['State_names','Gender','Amount_spent','Payment_method']].groupby(['State_names','Gender', 'Payment_method']).agg(['count', 'min', 'max'])
state_gender_res.head(12)
```

```python
df.groupby('Gender').mean()
```
