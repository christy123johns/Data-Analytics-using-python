# Scatter plot

```python
import pandas as pd
df= pd.read_excel('Online_Store_customer_data.xlsx')
df.head()
```

```python
import seaborn as sns
sns.scatterplot(x='Age',y='Amount_spent', data=df)
```

```python
sns.scatterplot(x='Age',y='Amount_spent', data=df, hue='Employees_status')
```
