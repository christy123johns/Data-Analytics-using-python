# Stacked Bar

```python
import pandas as pd
df= pd.read_excel('Online_Store_customer_data.xlsx')
df.head()
```

```python
pd.crosstab(df.Payment_method,df.Gender)
```

```python
pd.crosstab(df.Payment_method,df.Gender).plot.bar(stacked=True)
```

```python
pd.crosstab(df.Gender,df.Payment_method).plot.bar(stacked=True, color=['b','r','g'])
```

# 100% stacked bar

```python
pd.crosstab(df.Gender,df.Payment_method,normalize='index').plot.bar(stacked=True)
```

```python
pd.crosstab(df.Payment_method,df.Gender,normalize='index').plot.barh(stacked=True)
```

```python
x = pd.crosstab(df['Gender'], df['Payment_method'])
x.div(x.sum(1).astype(float), axis = 0).plot(kind = 'bar', stacked = True, figsize = (4, 4))
```
