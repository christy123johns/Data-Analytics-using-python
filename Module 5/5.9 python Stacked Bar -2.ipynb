# Stacked Bar

```python
import pandas as pd
df= pd.read_excel('HR.xlsx')
df.head()
```

```python
pd.crosstab(df.department,df.gender)
```

```python
pd.crosstab(df.department,df.gender).plot.bar(stacked=True)
```

```python
pd.crosstab(df.gender,df.department).plot.bar(stacked=True)
```

```python
pd.crosstab(df.gender,df.department,normalize='index').plot.bar(stacked=True)
```

```python
pd.crosstab(df.department,df.gender,normalize='index').plot.barh(stacked=True)
```

```python
DF = df.groupby(['gender'])['salary_2017','salary_2018'].mean()
DF = pd.DataFrame(DF).reset_index()
DF
```

```python
import matplotlib.pyplot as plt
plt.bar(DF['gender'], DF['salary_2018'])
plt.bar(DF['gender'], DF['salary_2017'], bottom=DF['salary_2018'])
plt.legend(['salary_2018', 'salary_2017'], title = "Legend", title_fontsize = 12, loc='upper left')

plt.title("Salary")

plt.xlabel('gender')
plt.ylabel('salary')

plt.show()
```
