# Boxplot

```python
import pandas as pd
df= pd.read_excel('Online_Store_customer_data.xlsx')
df.head()
```

```python
#We can plot this by one column or multiple columns. 
#For multiple columns, we need to pass columns name in y variable as a list.

df.plot(y=['Amount_spent'], kind='box')
```

```python
import seaborn as sns
sns.boxplot(x='Payment_method',y='Amount_spent', data=df, palette='ch:.25')
```

```python
import matplotlib.pyplot as plt
df.boxplot(by ='Employees_status', column =['Amount_spent'])
```

```python
sns.boxplot(x='Employees_status',y='Amount_spent', data=df)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(6,6)) 
df.boxplot(by ='Employees_status', column =['Amount_spent'],ax=ax, grid = False)
```
