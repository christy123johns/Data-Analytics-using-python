# Pie Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_excel('HR.xlsx')
df.head()
```

```python
df['department'].value_counts()
```

```python
colors= ['green','blue','black']
df['department'].value_counts(normalize=True).plot(kind='pie', color=colors)
```

```python
df['department'].value_counts(normalize=True).plot(kind='pie', autopct='%.2f%%')
```

```python
df['department'].value_counts(normalize=True).plot(kind='pie', autopct='%.2f%%')
plt.title('Department')
```

```python
import matplotlib.pyplot as plt
import numpy as np
#'r' - Red
#'g' - Green
#'b' - Blue
#'c' - Cyan
#'m' - Magenta
#'y' - Yellow
#'k' - Black
#'w' - White 


y = np.array([65.37, 30.34, 4.29])
mylabels = ['Research & Development','Sales', 'Human Resources']
mycolors = ["c", "y", "k"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.title('Departments')
plt.legend(title = "Departments")
plt.show()
```
