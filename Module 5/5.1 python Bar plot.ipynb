## Bar Plot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([5, 8, 3, 12])

plt.bar(x,y, color = "hotpink")
plt.title('Group members')
plt.show()
```
