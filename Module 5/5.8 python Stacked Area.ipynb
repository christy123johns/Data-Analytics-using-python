# Stacked Area

```python
import pandas as pd
df= pd.read_excel('Group performance.xlsx')
```

```python
import matplotlib.pyplot as plt
plt.stackplot(df['month'],df['team1'], df['team2'], df['team3'], labels=['team1','team2','team3'])
plt.legend(loc='upper left')

```

```python
import matplotlib.pyplot as plt
plt.stackplot(df['month'],df['team1'], df['team2'], df['team3'], labels=['team1','team2','team3'],  colors=['b','r','g'])
plt.legend(loc='upper right')

```
