# Pie Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_excel('Online_Store_customer_data.xlsx')
df.head()
```

```python
df['Payment_method'].value_counts()
```

```python
df['Payment_method'].value_counts(normalize=True).plot(kind='pie')
```

```python
df['Payment_method'].value_counts(normalize=True).plot(kind='pie', autopct='%.2f%%')
```

```python
colors= ['pink','hotpink','black']
df['Payment_method'].value_counts(normalize=True).plot(kind='pie', autopct='%.2f%%',colors=colors)
plt.title('Payment_method')
```

```python
import matplotlib.pyplot as plt
import numpy as np
#'r' - Red
#'g' - Green
#'b' - Blue
#'c' - Cyan
#'m' - Magenta
#'y' - Yellow
#'k' - Black
#'w' - White 


y = np.array([65.37, 30.34, 4.29])
mylabels = ['Research & Development','Sales', 'Human Resources']
mycolors = ["c", "y", "k"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.title('Departments')
plt.legend(title = "Department")
plt.show()
```
